# THREADING


```python
import threading 
import time
import numpy as np
```

```python
def sleeper(n,name):
    print("Hi, I am{}. Going to sleep for 5 seconds \n".format(name))
    time.sleep(n)
    print("{} has Woken Up ".format(name))
    
```

```python
t=threading.Thread(target=sleeper,name ="thread1",args=(5,"thread1"))
t.start()# when thread goes for sleeping it will start executing the main program
print("Hello")
print("Hello")
                   
```

```python
t=threading.Thread(target=sleeper,name ="thread1",args=(5,"thread1"))
t.start()
t.join()# puts a lock on thread to prvent it from switching  
print("Hello")
print("Hello")
```

# Daemon Thread


```python
total=4
def creator_1():
    global total
    for i in range(10):
        print("Incerease total in creator_1 \n")
        time.sleep(2)
        total=+1
        print("added item in 1\n")

def creator_2():
    global total
    for i in range(7):
        print("Increase total in creator_2 \n")
        time.sleep(1)
        total=+1
        print("added item in 2 \n")
    
def limiter():
    global total
    while True:
        if(total>5):
            print("Overloaded\n")
            total=total-3
            print("3 has been Subtracted \n")
        else:
            time.sleep(1)
            print('Waiting \n')       
```

```python
c1=threading.Thread(target=creator_1,args=())
c2=threading.Thread(target=creator_2,args=())
limit=threading.Thread(target=limiter,args=())
c1.start()
c2.start()
limit.start()
c1.join()
c2.join()
limit.join() # this gives the problem because we have used a infinite loop
print("The ending value of total is ", total)
```

```python
c1=threading.Thread(target=creator_1,args=())
c2=threading.Thread(target=creator_2,args=())
limit=threading.Thread(target=limiter,args=(),daemon=True)
print(limit.isDaemon())
c1.start()
c2.start()
limit.start()
c1.join()
c2.join()
limit.join()
print("The ending value of total is ", total)
```

# LOCKS IN THREADING


```python
'''This example is executing without locks 
See the output of the code
expected ouput should be 0 because we are adding 5 to x and subtracting 5 from x
but the output is not same as the expected ouput
'''


x=0
count=100000
def adding_2():
    global x
    for i in range(count):
        x=x+2
def adding_3():
    global x
    for i in range(count):
        x=x+3

def subtracting_4():
    global x
    for i in range(count):
        x=x-4
def subtracting_1():
    global x
    for i in range(count):
        x=x-1
        
t1=threading.Thread(target=adding_2)
t2=threading.Thread(target=adding_3)
t3=threading.Thread(target=subtracting_4)
t4=threading.Thread(target=subtracting_1)

t1.start()
t3.start()
t2.start()
t4.start()

t1.join()
t3.join()
t2.join()
t4.join()

print("The value of x: ", x)
```

```python
'''this is same example 
but using locks and 
anyalyze the ouput
'''


x=0
count=100000
lock=threading.Lock()
def adding_2():
    global x
    with lock:
        for i in range(count):
            x=x+2
def adding_3():
    global x
    with lock:
        for i in range(count):
            x=x+3

def subtracting_4():
    global x
    with lock:
        for i in range(count):
            x=x-4
def subtracting_1():
    global x
    with lock:
        for i in range(count):
            x=x-1
        
t1=threading.Thread(target=adding_2)
t2=threading.Thread(target=adding_3)
t3=threading.Thread(target=subtracting_4)
t4=threading.Thread(target=subtracting_1)

t1.start()
t3.start()
t2.start()
t4.start()

t1.join()
t3.join()
t2.join()
t4.join()

print("The value of x: ", x)
```
